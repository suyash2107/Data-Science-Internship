##Normalization

```python
x=np.array([11,45,23,17,21,15,7,33])
```

```python
def normalize(a):
    norm = (a - a.min()) / (a.max() - a.min())
    return norm
```

```python
normalize (x)
```

```python
nx=normalize(x)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.plot(x)
plt.subplot(1,2,2)
plt.title('Normalize Data')

plt.plot(nx);

```

```python
y=np.array([678,545,376,487,532,856,465,378,366,754,734])
ny=normalize(y)
ny
```
