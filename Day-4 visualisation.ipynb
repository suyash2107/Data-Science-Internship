### BOX PLOT

```python
marks=[67,63,70,71,72,56,51,21,95]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks,75)
q1=np.percentile(marks,25)
```

```python
 q3,q1
```

```python
iqr=q3-q1
iqr   ##interquartile range
```

```python
#lower bound
lb= q1-1.5*iqr
lb
```

```python
#upper bound
ub= q3+1.5*iqr
ub
```

```python
lb,ub
```

# SEA BORN

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python


sns.jointplot(x='tip',y='total_bill',data=tips,hue='time');
plt.show()

```

## HISTOGRAM

```python
plt.xlabel('tip Amout')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)
plt.show()
```

```python
plt.title('outliers in tip')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxenplot(x='day',y='tip',
             data=tips);
plt.show()
```

```python
sns.pairplot(tips,hue='smoker');
plt.show()
```
